```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### **Heuristíca Gananciosa:**

```python
fc = pd.read_csv("./../projeto/resultados/gananciosa_filmes_categorias.csv")
fm = pd.read_csv("./../projeto/resultados/gananciosa_medias.csv")
```

```python
fc
```

Filmes totais x Filmes vistos

```python
plt.figure(figsize=(10, 5))
for i in range(2, 16,2):
    fxf = fc[fc["n_categorias"] == i]
    plt.plot(fxf["n_filmes"], fxf["v_filmes"], label=str(i))
plt.legend(title="categorias") 
plt.title('Filmes x Filmes (Gananciosa)')
plt.xlabel('número de filmes')
plt.ylabel('filmes assistidos')
```

```python
plt.figure(figsize=(10, 5))
for i in range(1, 50,6):
    fxf = fc[fc["n_categorias"] == i]
    plt.plot(fxf["n_filmes"], fxf["t_execucao"], label=str(i))
plt.legend(title="categorias") 
plt.title('Filmes x Tempo de Execução (Gananciosa)')
plt.xlabel('número de filmes')
plt.ylabel('tempo[s]')
```

### Comparação Gananciosa x Aleatória

```python
fcg = pd.read_csv("./../projeto/resultados/gananciosa_filmes_categorias.csv")
fmg = pd.read_csv("./../projeto/resultados/gananciosa_medias.csv")
fca = pd.read_csv("./../projeto/resultados/aleatoria_filmes_categorias.csv")
fma = pd.read_csv("./../projeto/resultados/aleatoria_medias.csv")
```

Filmes totais x Filmes vistos

```python
plt.figure(figsize=(10, 5))

for i in range(1, 50, 6):
    a = fca[fca["n_categorias"] == i]
    g = fcg[fcg["n_categorias"] == i]
    plt.plot(g["n_filmes"], g["v_filmes"], label="Gananciosa", color = "red")
    plt.plot(a["n_filmes"], a["v_filmes"], label="Aleatória", color = "blue")
plt.legend(["Gananciosa", "Aleatória"],title="heurísticas") 
plt.title('Filmes x Filmes')
plt.xlabel('número de filmes')
plt.ylabel('filmes assistidos')
```

Categorias x Filmes assistidos

```python
plt.figure(figsize=(10, 5))

for i in range(1000, 50000,6000):
    a = fca[fca["n_filmes"] == i]
    g = fcg[fcg["n_filmes"] == i]
    plt.plot(g["n_categorias"], g["v_filmes"], label="Gananciosa", color = "red")
    plt.plot(a["n_categorias"], a["v_filmes"], label="Aleatória", color = "blue")
plt.legend(["Gananciosa", "Aleatória"],title="heurísticas") 
plt.title('Categorias x Filmes')
plt.xlabel('número de categorias')
plt.ylabel('filmes assistidos')
```

Filmes x Tempo de Execução

```python
plt.figure(figsize=(10, 5))

for i in range(1, 50, 6):
    a = fca[fca["n_categorias"] == i]
    g = fcg[fcg["n_categorias"] == i]
    plt.plot(g["n_filmes"], g["t_execucao"], label="Gananciosa", color = "red")
    plt.plot(a["n_filmes"], a["t_execucao"], label="Aleatória", color = "blue")
plt.legend(["Gananciosa", "Aleatória"],title="heurísticas") 
plt.title('Filmes x Tempo de Execução')
plt.xlabel('número de filmes')
plt.ylabel('tempo[s]')
```
